```python
import re
```

```python
lst="My name is Varun and I am 44 years old. IO My year of birth is 1981 and day is 20"
y=re.findall('[0-9]+',lst)
print(y)

y=re.findall('[AEIOU]+',lst)
print(y)

```

```python
lst="From: Vaurn Sharma : mail body"
y=re.findall('^F.+:',lst)
print(y)

y=re.findall('^F.+?:',lst)
print(y)
```

```python
mailbosy="Mail received from: Varun.Sharma@outlook.com"

email=re.findall(r'@(\S+)',mailbosy)
print(email)
```

```python

filename="D:\\regex_sum_2435527.txt"
fl=open(filename,"r")
numberlist=[]

for ln in fl:
    numberlist.append(re.findall('[0-9]+',ln))

fl.close()
sumTotal=0
for listitem in numberlist:
    for number in listitem:
        sumTotal=int(number)+sumTotal
print(sumTotal)
```

```python
import socket
mysocket=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
mysocket.connect(('google.com',80  ))
cmd='GET http://www.google.com HTTP/1.0\n\n'.encode()
mysocket.send(cmd)

while True:
    data=mysocket.recv(512)
    if len(data)<1:
        break
    print(data.decode(),end='')
    
mysocket.close()
```
