```python
import re
```

```python
lst="My name is Varun and I am 44 years old. IO My year of birth is 1981 and day is 20"
y=re.findall('[0-9]+',lst)
print(y)

y=re.findall('[AEIOU]+',lst)
print(y)

```

```python
lst="From: Vaurn Sharma : mail body"
y=re.findall('^F.+:',lst)
print(y)

y=re.findall('^F.+?:',lst)
print(y)
```

```python
mailbosy="Mail received from: Varun.Sharma@outlook.com"

email=re.findall(r'@(\S+)',mailbosy)
print(email)
```

```python

filename="D:\\regex_sum_42.txt"
fl=open(filename,"r")
numberlist=[]

for ln in fl:
    numberlist.append(re.findall('[0-9]+',ln))

fl.close()
sumTotal=0
for listitem in numberlist:
    for number in listitem:
        sumTotal=int(number)+sumTotal
print(sumTotal)
```
